From this, it can be infered that, the dataset has no null values in it. To check this we can use the following function.


### Finding null values in the data


```python
data.isnull().sum()
```

### Unique wind speed values


```python
data["Wind Speed_km/h"].unique()
```

### Using Group by to show records when the weather is "exactly clear"

```python
data.groupby('Weather').get_group('Clear')
```

### Number of occurences of each distinct value in the Weather column

```python
data.value_counts('Weather')
```

### Renaming the column 'Weather" to "Weather Condition"

```python
data.rename(columns = {'Weather':'Weather Condition'})
```

### Mean of "Visibility column"

```python
data.Visibility_km.mean()
```

### Standard deviation of "Pressure column"

```python
data.Press_kPa.std()
```

### Variance of "relative humidity"

```python
data["Rel Hum_%"].var()
```

### Instances where exactly "snow" was recorded using Filtering


```python
data[data['Weather']=='Snow']
```

### Instances where "snow" was recorded using str.contains

```python
data[data['Weather'].str.contains('Snow')]
```

### Finding instances when windspeed is above 24 and visibility = 25

```python
data[(data["Wind Speed_km/h"]>24)& (data["Visibility_km"] == 25)]
```

### Mean of each column against each 'Weather' using group by

```python
data.groupby('Weather').mean()
```

### Minimum value of each column against each Weather

```python
data.groupby('Weather').min()
```

### Maximum value of each column against each Weather

```python
data.groupby('Weather').max()
```

### Relationship between Temperature and Pressure 

```python
sns.scatterplot(x="Temp_C",y="Press_kPa",data = data)
```

### Top 10 Weather Categories

```python
top_n = 10  


top_categories = data['Weather'].value_counts().nlargest(top_n).index

mp.figure(figsize=(10, 6))
sns.countplot(data=data, y='Weather', order=top_categories)

mp.show()
```
